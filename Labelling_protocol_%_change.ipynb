```python
dfs = [df, df1, df2]
df_final = pd.concat(dfs)
df_final.head()
```

```python
df_final.dropna(axis = 0, how = 'any', inplace = True)
df_final.info()
```

We are going to recalculate the label by looking at the relative change in trends. If relative change > 1, or < 1 then we are going to consider them increases/decreases.

```python
x = df_final['PrevAvgPriceChange'].values
y = df_final['PostAvgPriceChange'].values
```

```python
labs = []

for i in range(len(x)):
    if abs(y[i]-x[i])/abs(x[i]) > 1:
        if y[i] > x[i]:
            labs.append('increase')
        else:
            labs.append('decrease')
    else:
        labs.append('neutral')
```

```python
df_final['Label'] = labs
```

```python
df_final.head()
```

```python
import seaborn as sns
sns.set_style('darkgrid')
incr_counts = 0
decr_counts = 0
neutral_counts = 0

for i in df_final.Label:
    if i == 'neutral':
        neutral_counts += 1
    elif i == 'increase':
        incr_counts += 1
        
decr_counts = len(df_final) - (incr_counts + neutral_counts)

counts = [neutral_counts, incr_counts, decr_counts]
labels = ['Neutral', 'Increase', 'Decrease']

plt.bar(labels, counts, color = 'lightsteelblue')
plt.ylim([0, 700])
plt.show()

"""
print('Price increased: {} times'.format(incr_counts))
print('Price decreased: {} times'.format(decr_counts))
print('Price did not move: {} times'.format(neutral_counts))"""
```
